## Drawing error terms recursively

Here, I am going to describe in very rough terms how to go about drawing error terms recursively so that things are consistent with the "unilateral deviation" story that we are using. These error terms can then be applied in just about any way possible. 

A key thing to realize is that the presence of lags make it impossible to draw these terms "all at once" and in fact it works best to do them backwards recursively. So, under the assumption that what everyone else is doing is fixed, we want to know - what are some plausible values for unoberved actions of a given player? What alternative viewerships and prices are consistent with the idea that our man has chosen a Nash equilibrium profile? 



We start at the end (at time $t=6$ in our model). Suppose that our station has chosen action $b_6$ in period $6$, and let $b'_6$ denote the alternative strategy available to the player (i.e., if the player chose to actually broadcast local news, the prime would represent entertainment. Let's start by first computing hypothetical viewership and drawing a viewership error term for the utility index:

$U_{vi6}=X_{i6}\beta'+u_{m6}+u_i+\epsilon'_{v6}$

We then run this through the multinomial logit to generate a hypothetical share:

$v'_{i6}$. 

Now, we know that all strategies should generate less revenue than the observed one. This is also true of the strategy that follows exactly the observed strategy but only deviates in the last period. That is:

$
\sum_{t=1}^5p_{it} +p'_{i6}\leq \sum_{t=1}^5p_{it} +p_{i6}
$

Or

$
p'_{i6} \leq p_{i6}
$

Or

$
X\beta_p+\ln v'_{i6}+\epsilon_{p6}'< \ln p_{i6}
$

Or

$
\epsilon_{p6}'< \ln p_{i6}-X\beta_p- \ln v'_{i6}
$

This last expression can be used to draw a value for $\epsilon_{p6}'$ from a truncated normal distribution. This reflects the fact that by assumption the strategy of following everything as observed until the last period and then deviating can't give a higher price - either because of viewership or unobserved attributes of price - than what we see. 

Back up to period 5. There are now a couple of different approaches that the station could take. It could:

1. Play $b'_5$ and then $b_6$, or
2. Play $b_'5$ and then play $b_6'$

By assumption, neither or these can generate a profit higher than what we observe. We don't have to worry about the leading 4 periods, as those prices will be the same. What we need to make sure is true, however, is that:

$
e^{\ln p^{(1)}_5}+e^{\ln p^{(1)}_6} \leq p_5+p_6
$

and

$
e^{\ln p^{(2)}_5}+e^{\ln p^{(2)}_6} \leq p_5+p_6
$


There is a subtle problem with just going ahead and doing this. We note that in many permutations, viewership that we obtain in period $5$ will impact viewership in period $6$, and the model will constrain exactly how large viewership can be. 

Let's rearrange the above so that we get:

$
e^{\ln p^{(1)}_5} \leq p_5+p_6-e^{\ln p^{(1)}_6}
$

The problem is that, if we aren't careful, we can get a negative-left hand term, which will make taking logs impossible at some point down the line. The good news is that this possibility actually provides an additional secret constraint. 

$
e^{\ln p^{(2)}_5} \leq p_5+p_6-e^{\ln p^{(2)}_6}
$

We write the two above as:

$
\ln p^{(1)}_5 \leq \ln( p_5+p_6-e^{\ln p^{(1)}_6})
$

$
\ln p^{(2)}_5 \leq \ln(p_5+p_6-e^{\ln p^{(2)}_6})
$

Now, it is clear that if the best continuation strategy out of $(1)$ or $(2)$ generates less of a payoff than observed, then so will the other one. Also, both strategies are monotone in the unobserved error term, so we need only pick out the one with the higher mean. That is, choose the maximum of:

$
\max_{(i)}[ e^{X\beta_p'+\eta\ln v'_5+e_5}+p'_6]
$

Now, here is where we can run into trouble. Suppose we find the best continuation strategy from the above. We now need to set things up to draw a truncated normal random variable. Rearranging, we get:

$
X\beta_p+\eta' \ln v'_5 +\epsilon_{p5} \leq p_5+p_6 - p'_6
$

Or

$
\epsilon_{p5} \leq \ln (p_5+p_6 - p'_6-X\beta-\eta'\ln v'_5)
$

Now, we might wind up with a negative number on the right-hand side. While this might be considered a numerical artifact, it could also be considered as a constraint suggesting that whatever error term we used in the viewership equation is generating implausibly high viewerships. So, we might build this in as suggesting an upper bound on the viewership term, as follows:

$
\ln v'_5 \leq \frac{p_5+p_6 - p'_6-X\beta}{\eta'}
$

Now, let's unpack this a bit to get, in terms of shares: 

$
\ln s'_5+\ln M \leq \frac{p_5+p_6 - p'_6-X\beta}{\eta'}
$

Or

$
\ln s'_5\leq \frac{p_5+p_6 - p'_6-X\beta}{\eta'}-\ln M
$

Note that the largest possible value of the above is:

$
\ln s'_5=\frac{p_5+p_6 - p'_6-X\beta}{\eta'}-\ln M
$

We could solve this for $\epsilon$ and get the upper bound, and then go from there. That would seem to work. Let's try it after lunch...
